```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the JSON file
with open('dat/QSP_sim_2.json', 'r') as file:
    data = json.load(file)

# Extract the arrays
predictions = np.array(data['y_sim'])
actuals = np.array(data['y_actual'])

# Find the average difference between the actual and simulated
diff = predictions - actuals

print(predictions)
print(actuals)
print(diff)
```

```python

phi = []
diffs = []

for i in range(2, 12, 2):

    # Load the JSON file

    f = 'dat/QSP_sim_' + str(i) + '.json'

    with open(f, 'r') as file:
        data = json.load(file)

    # Extract the arrays
    predictions = np.array(data['y_sim'])
    actuals = np.array(data['y_actual'])

    # Find the average difference between the actual and simulated
    diffs.append(np.mean(abs(np.absolute(predictions) - np.absolute(actuals))))
    phi.append((i-1)*2 + 1)

print(diffs)

```

```python
# Verify that values make sense
print(phi)
print(diffs)

plt.scatter(phi, diffs, color='blue', marker='o')

# Add labels and title
plt.xlabel('Circuit depth')
plt.ylabel('Average pointwise error')
plt.title('Average error of simulated estimate for sin(x) using PSA (100 samples/point)')

plt.xticks(range(3, 21, 1))

# Show the plot
plt.show()
```
