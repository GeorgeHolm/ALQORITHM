```python
from pyqsp.angle_sequence import QuantumSignalProcessingPhases
from pyqsp.response import PlotQSPResponse
from pyqsp.poly import StringPolynomial
from numpy.polynomial import Polynomial

# installed tensorflow, seaborn, pyqsp, qiskit_ibm_runtime onto default env
```

```python
poly_1 = Polynomial([-0.2,0,1.1])
```

```python
ang_seq_1 = QuantumSignalProcessingPhases([-0.2,0,1.1], signal_operator="Wx")
print(ang_seq_1)
```

```python
# 1.1x^2 - 0.2
PlotQSPResponse(ang_seq_1, target=poly_1, signal_operator="Wx")
```

```python
# Warning: Time consuming [~ 1 min] since uses more accurate tensorflow to compute phis
poly = StringPolynomial("np.sin(3*x)", 7)
ang_seq = QuantumSignalProcessingPhases(poly, signal_operator="Wx", method="tf")
```

```python
# sin(x)
PlotQSPResponse(ang_seq[:,0], target=poly, signal_operator="Wx")
```
